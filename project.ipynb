```python
x= df[['MW','T_nb( C )','specific gravity','dipole moment']]
y=df[['Lower explosion(%)','Upper explosion(%)','Toxicity(mg/L)(inhalation)','time(h)(toxicity)','Dermal(mg/kg)','Lower flamebility(%)','upper flamebility(%)']]
```

```python
x_train, x_test, y_train,y_test = train_test_split(x,y,random_state=2)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_train_pred=model.predict(x_train)
y_test_pred=model.predict(x_test)
```

```python
mse = model.score(x_train, y_train)
print ( f'mse : {mse: .2f}')
```

```python
mse = model.score(x_test, y_test)
print ( f'mse : {mse: .2f}')
```
